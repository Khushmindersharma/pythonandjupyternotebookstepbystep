```python
# Matplotlib is both 2d plotting library.

import matplotlib.pyplot as plt

# Linear Graph.

x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)

plt.show() # this function help in removing the extra data on the graph.
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r' # this help in coloring the line in the graph.

plt.plot(x,y,c)

plt.show() # this function help in removing the extra data on the graph.
```

```python
# Bar Graph.

x = [1,2,3,4]
y = [5,6,7,8]



plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']


plt.bar(x,y, color = c)
plt.show()
```

```python
# bar graph using arrays.

import numpy as np

x = np.array(["a","b","c","d"])
y = np.array([3,8,1,10])

plt.barh(x,y, color ="hotpink") # color code will also work in this process.
plt.show()
```

```python
x = np.array(["a","b","c","d"])
y = np.array([3,8,1,10])

plt.bar(x,y, color ="hotpink", width = 0.1) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'girl.jpg'

# Opening image using PIL

image = Image.open(fname).convert('L')

# mapping image to gray scale.

plt.imshow(image, cmap = 'inferno')
plt.title("Beautiful Girl")
plt.xlabel("350 Length")
plt.ylabel("200 Breath")
plt.grid()
plt.show()
```

```python
# Matplotlib - pie chart

x = [10,20,30,40]
y = ['English','Hindi','Maths','Science']

plt.pie(x, labels = y)
plt.show()

```

```python
x = [10,20,30,40]
y = ['English','Hindi','Maths','Science']
c = ['Yellow','Magenta','Brown','Aqua']

plt.pie(x, labels = y, colors = c)
plt.legend()
plt.show()
```
